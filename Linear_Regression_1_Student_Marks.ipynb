```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# Create Test and Train variables using test_train_split from sklearn library:
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2,random_state=0)   # test size = 20%, train = 80% of data
print(x_test.shape,x_train.shape,y_test.shape,y_train.shape)
```

```python
print(x)
```

```python
len(x_train)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
model_predictions = model.predict(x_test)
```

```python
model.predict(x_test)
```

```python
y_pred = model.predict(x_test)
print('Mean squared error: %2f'%mean_squared_error(y_test,y_pred))

# It means close estimates or forecast is to actual values. The lower value indicates a better fit.
```

```python
print('Coefficient of determination:%2f'%r2_score(y_test,y_pred))
# In this model R value is small so it means 90% of variance of dependent variables are studied and explained by the variance of independent variables.
# Regression model fits the data well if the distance between the observations and the predicted values are small and unbiased.
# so, whatever predictions will be done will be 93% accurate.
```

```python
coef=model.coef_
model.coef_
# It indicates that the value of idependent variables (x) increases so the mean of dependent variable(y) also increase.
```

```python
model.intercept_
```

```python
model.score(x_test,y_test)
# It means the observed and the predicted outcomes values are highly correlated.
```
